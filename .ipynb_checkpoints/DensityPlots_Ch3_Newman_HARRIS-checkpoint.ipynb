Example 3.1 in Newman (Wave Interference)

```python
import numpy as np
import math
import matplotlib.pyplot as plt

wavelength = 5.0
k = (2*np.pi)/wavelength
xiO = 1.0
separation = 20.0                       #Separation of centers in cm
side = 100.0                            #Side of squares in cm
points = 500                            #Number of grid points along each side 
spacing = side / points                 #Spacing of points in cm

#Calculated the positions of the centers of the circles
x1 = (side / 2) + (separation / 2)
y1 = (side / 2) 
x2 = (side / 2) - (separation / 2)
y2 = (side / 2) 

#made an empty array to store the height values
xi = np.empty ([points,points], float)

#Calculated the values that go into the array
for i in range (points):
    y = spacing*i
    for j in range (points):
        x = spacing*j
        amplitude = 1
        r1 = np.sqrt((x-x1)**2 + (y-y1)**2)
        r2 = np.sqrt((x-x2)**2 + (y-y2)**2)
        xi [i,j] = (amplitude)*np.sin(k*r1) + (amplitude)*np.sin(k*r2)

#Made the plot
fig, ax = plt.subplots()
im = ax.imshow(xi, origin = "lower", extent = [0, side, 0, side])

ax.set_title('Wave Interference')
plt.gray()
plt.show()
    
```

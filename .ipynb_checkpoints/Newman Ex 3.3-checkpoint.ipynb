Exercise 3.3 in Newman on STM (scanning tunneling microscopy)

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("stm.txt", float)     
data = data [:]                              #Load all the data from the txt file into an array

#Made the plot of the data
fig, ax = plt.subplots()                     
im = ax.imshow(data, origin = "lower")

ax.set_title('STM of Silicon Surface')
plt.gray()
plt.show()
```
